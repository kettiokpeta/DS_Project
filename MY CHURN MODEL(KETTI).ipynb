### RFM-analysis
**First, calculate RFM metrics**

```python
# filtering neccessary columns

RFM = trans_data[['Customer_ID', 'Invoice_NO', 'Revenue', 'Order_Date']]
RFM.head()
```

```python
# get last transaction date

lastTransaction = max(mydata.Order_Date)
lastTransaction
```

so our last transaction is 28th Janurary 2014

```python
# Aggregate the Data for RFM analysis

RFM = RFM.groupby('Customer_ID').agg({'Order_Date': lambda x: (lastTransaction - x.max()).days, 'Invoice_NO' : 'count', 'Revenue' : 'sum'})
RFM.head()
```

```python
# Rename the columns

RFM.rename(columns = {'Order_Date' : 'Recency', 
                      'Invoice_NO' : 'Frequency', 
                      'Revenue' : 'Monetary'}, inplace = True)
RFM.head()
```

As the three columns are grouped by customers and count the days from the max date value, Recency is the days since the last purchase of a customer. Frequency is the number of purchases of a customer and Monetary is the total amount of spend of a customer.

### RFM quartiles

We will group the customers based on Recency and Frequency. We will use quantile values to get three equal percentile groups an then make three separate gruops. As the lower Recency value is the better, we will label them in decreasing order.

```python
quantiles = RFM.quantile(q=[0.25,0.5,0.75])
quantiles = quantiles.to_dict()
```

```python
segmented_RFM = RFM.copy()
```

```python
def RScore(x,p,d):
    if x <= d[p][0.25]:
        return 1
    elif x <= d[p][0.50]:
        return 2
    elif x <= d[p][0.75]: 
        return 3
    else:
        return 4
    
def FMScore(x,p,d):
    if x <= d[p][0.25]:
        return 4
    elif x <= d[p][0.50]:
        return 3
    elif x <= d[p][0.75]: 
        return 2
    else:
        return 1
```

```python
segmented_RFM['R_quartile'] = segmented_RFM['Recency'].apply(RScore, args=('Recency',quantiles,))
segmented_RFM['F_quartile'] = segmented_RFM['Frequency'].apply(FMScore, args=('Frequency',quantiles,))
segmented_RFM['M_quartile'] = segmented_RFM['Monetary'].apply(FMScore, args=('Monetary',quantiles,))

segmented_RFM.head()
```

* Best Recency score = 1: Purchased very recently.
* Best Frequency score = 1: highest quantity purchases.
* Best Monetary score = 1: spent the most.

```python
segmented_RFM['RFM_segment'] = segmented_RFM.R_quartile.map(str) + segmented_RFM.F_quartile.map(str) + segmented_RFM.M_quartile.map(str)
segmented_RFM['RFM_Score'] = segmented_RFM[['R_quartile', 'F_quartile', 'M_quartile']].sum(axis = 1)

segmented_RFM.head()
```

```python
segmented_RFM['RFM_Score'].unique()
```

```python
# drop RFM quartile columns since we have segments

segmented_RFM.drop(['R_quartile', 'F_quartile', 'M_quartile'], axis=1, inplace = True)
segmented_RFM.head()
```

```python
# calculate averge values for each RFM

RFM_agg = segmented_RFM.groupby('RFM_Score').agg({
    'Recency' : 'mean',
    'Frequency' : 'mean',
    'Monetary' : ['mean', 'count']})

RFM_agg.round(1).head()
```

```python
# assign labels from total score

score_labels = ['Gold', 'Silver', 'Bronze', 'Green']
score_groups = pd.qcut(segmented_RFM.RFM_Score, q = 4, labels = score_labels)
segmented_RFM['RFM_Level'] = score_groups.values
segmented_RFM.head()
```

```python
segmented_RFM['RFM_segment'].nunique()
```

```python
# Retrieve the percentage % of the most valuable customers (Those in 111 segment)

print("Our most valuable customers of 111 segment is {:.2f}%".format(len(segmented_RFM[segmented_RFM['RFM_segment'] == '111'])/len(segmented_RFM)*100))
```

### Customer Segmentation with Kmeans

#### K-means gives the best result under the following conditions:
* Datas distribution is not skewed.
* Data is standardised (i.e. mean of 0 and standard deviation of 1).

```python
def check_skew(df_skew, column):
    skew = stats.skew(df_skew[column])
    skewtest = stats.skewtest(df_skew[column])
    plt.title('Distribution of ' + column)
    sns.distplot(df_skew[column])
    print("{}'s: Skew: {}, : {}".format(column, skew, skewtest))
    return
```

```python
# Plot all 3 graphs together for summary findings
plt.figure(figsize=(9, 9))

plt.subplot(3, 1, 1)
check_skew(RFM,'Recency')

plt.subplot(3, 1, 2)
check_skew(RFM,'Frequency')

plt.subplot(3, 1, 3)
check_skew(RFM,'Monetary')

plt.tight_layout()
plt.savefig('before_transform.png', format='png', dpi=1000)
```

The data is skewed, therefore we will perform log transformations to reduce the skewness of each variable.
I added a small constant as log transformation demands all the values to be positive.

```python
# define function for the values below 0

def neg_to_zero(x):
    if x <= 0:
        return 1
    else:
        return x
```

```python
# apply the function to Recency and Monetary Value column

segmented_RFM['Recency'] = [neg_to_zero(x) for x in segmented_RFM.Recency]
segmented_RFM['Monetary'] = [neg_to_zero(x) for x in segmented_RFM.Monetary]
segmented_RFM.head()
```

```python
# unskew the data

RFM_log = segmented_RFM[['Recency', 'Frequency', 'Monetary']].apply(np.log, axis = 1).round(3)
RFM_log.head()
```

```python
RFM_log.describe()
```

```python
# scale the data

scaler = StandardScaler()
RFM_scaled = scaler.fit_transform(RFM_log)
```

```python
# transform into a dataframe

RFM_scaled = pd.DataFrame(RFM_scaled, index = RFM.index, columns = RFM_log.columns)
RFM_scaled.head()
```

```python
# plot the distribution of RFM values

plt.figure(figsize=(9, 9))

plt.subplot(3, 1, 1)
check_skew(RFM_scaled,'Recency')

plt.subplot(3, 1, 2)
check_skew(RFM_scaled,'Frequency')

plt.subplot(3, 1, 3)
check_skew(RFM_scaled,'Monetary')

plt.tight_layout()
plt.savefig('after_transform.png', format='png', dpi=1000)
```

### K-Means Clustering

With the Elbow method, we can get the optimal number of clusters.

```python
# initiate an empty dictionary

wcss = {}

# Elbow method with for loop

for k in range(1, 11):
    kmeans = KMeans(n_clusters= k, init= 'k-means++', max_iter= 300)
    kmeans.fit(RFM_scaled)
    wcss[k] = kmeans.inertia_
```

```python
# plot the WCSS values

sns.pointplot(x = list(wcss.keys()), y = list(wcss.values()))
plt.xlabel('K Numbers')
plt.ylabel('WCSS')
plt.show()
```

The optimal K value is found to be 4 using the elbow method

```python
# choose n_clusters = 4 
clus = KMeans(n_clusters= 4, init= 'k-means++', max_iter= 300)
clus.fit(RFM_scaled)
```

```python
# Assign the clusters to datamart

segmented_RFM['K_Cluster'] = clus.labels_
segmented_RFM.head()
```

### COSTUMER TENURE

```python
# Filter columns for tenure

tenure = trans_data[['Customer_ID', 'Order_Date']]
tenure = tenure.groupby('Customer_ID').agg({'Order_Date':['min', 'max']})
tenure.columns = ["_".join(tenure) for tenure in tenure.columns.ravel()]
tenure.head()
```

```python
# Tenure in Days, months and years

tenure['Tenure_Days'] = tenure['Order_Date_max'] - tenure['Order_Date_min']
tenure['Tenure_Months'] = tenure['Tenure_Days'] / np.timedelta64(1, 'M')
tenure['Tenure_Y'] = tenure['Tenure_Days'] / np.timedelta64(1, 'Y')

def yearmonthdiff(row):
    s = row['Order_Date_max']
    e = row['Order_Date_min']
    y = s.year - e.year
    m = s.month - e.month
    d = s.day - e.day
    if m < 0:
        y = y - 1
        m = m + 12
    if m == 0:
        if d < 0:
            m = m -1
        elif d == 0:
            s1 = s.hour*3600 + s.minute*60 + s.second
            s2 = e.hour*3600 + e.minute*60 + e.second
            if s1 < s2:
                m = m - 1
    return '{}y, {}m'.format(y,m)

tenure['Tenure_years'] = tenure.apply(yearmonthdiff, axis=1)

tenure.head()
```

```python
# Split the customers into 4 groups of tenure Months for visualization

tenure_0m = tenure.Tenure_Months[(tenure.Tenure_Months < 10) & (tenure.Tenure_Months >= 0)]
tenure_10m = tenure.Tenure_Months[(tenure.Tenure_Months < 20) & (tenure.Tenure_Months >= 10)]
tenure_20m = tenure.Tenure_Months[(tenure.Tenure_Months < 30) & (tenure.Tenure_Months >= 20)]
tenure_30mabove = tenure.Tenure_Months[tenure.Tenure_Months >= 30]

x = ["0-9 months","10-19 months","20-29 months","30 months & above"]
y = [len(tenure_0m.values),len(tenure_10m.values),len(tenure_20m.values),len(tenure_30mabove.values)]

plt.figure(figsize=(10,6))
sns.barplot(x=x, y=y, palette="rocket")
plt.title("Customer Tenure")
plt.xlabel("Tenue by Group Months")
plt.ylabel("Number of Customer")
plt.show()
```

```python
def calculate_vif(features):
    vif = pd.DataFrame()
    vif["Features"] = features.columns
    vif["VIF"] = [variance_inflation_factor(features.values, i) for i in range(features.shape[1])]    
    return(vif)

calculate_vif(features)
```

### Information value

## WOE and IV

```python
import pandas.core.algorithms as algos
from pandas import Series
import scipy.stats.stats as stats
import traceback

max_bin = 20
force_bin = 3

# define a binning function
def mono_bin(Y, X, n = max_bin):
    
    df1 = pd.DataFrame({"X": X, "Y": Y})
    justmiss = df1[['X','Y']][df1.X.isnull()]
    notmiss = df1[['X','Y']][df1.X.notnull()]
    r = 0
    while np.abs(r) < 1:
        try:
            d1 = pd.DataFrame({"X": notmiss.X, "Y": notmiss.Y, "Bucket": pd.qcut(notmiss.X, n)})
            d2 = d1.groupby('Bucket', as_index=True)
            r, p = stats.spearmanr(d2.mean().X, d2.mean().Y)
            n = n - 1 
        except Exception as e:
            n = n - 1

    if len(d2) == 1:
        n = force_bin         
        bins = algos.quantile(notmiss.X, np.linspace(0, 1, n))
        if len(np.unique(bins)) == 2:
            bins = np.insert(bins, 0, 1)
            bins[1] = bins[1]-(bins[1]/2)
        d1 = pd.DataFrame({"X": notmiss.X, "Y": notmiss.Y, "Bucket": pd.cut(notmiss.X, np.unique(bins),include_lowest=True)}) 
        d2 = d1.groupby('Bucket', as_index=True)
    
    d3 = pd.DataFrame({},index=[])
    d3["MIN_VALUE"] = d2.min().X
    d3["MAX_VALUE"] = d2.max().X
    d3["COUNT"] = d2.count().Y
    d3["EVENT"] = d2.sum().Y
    d3["NONEVENT"] = d2.count().Y - d2.sum().Y
    d3=d3.reset_index(drop=True)
    
    if len(justmiss.index) > 0:
        d4 = pd.DataFrame({'MIN_VALUE':np.nan},index=[0])
        d4["MAX_VALUE"] = np.nan
        d4["COUNT"] = justmiss.count().Y
        d4["EVENT"] = justmiss.sum().Y
        d4["NONEVENT"] = justmiss.count().Y - justmiss.sum().Y
        d3 = d3.append(d4,ignore_index=True)
    
    d3["EVENT_RATE"] = d3.EVENT/d3.COUNT
    d3["NON_EVENT_RATE"] = d3.NONEVENT/d3.COUNT
    d3["DIST_EVENT"] = d3.EVENT/d3.sum().EVENT
    d3["DIST_NON_EVENT"] = d3.NONEVENT/d3.sum().NONEVENT
    d3["WOE"] = np.log(d3.DIST_EVENT/d3.DIST_NON_EVENT)
    d3["IV"] = (d3.DIST_EVENT-d3.DIST_NON_EVENT)*np.log(d3.DIST_EVENT/d3.DIST_NON_EVENT)
    d3["VAR_NAME"] = "VAR"
    d3 = d3[['VAR_NAME','MIN_VALUE', 'MAX_VALUE', 'COUNT', 'EVENT', 'EVENT_RATE', 'NONEVENT', 'NON_EVENT_RATE', 'DIST_EVENT','DIST_NON_EVENT','WOE', 'IV']]       
    d3 = d3.replace([np.inf, -np.inf], 0)
    d3.IV = d3.IV.sum()
    
    return(d3)

def char_bin(Y, X):
        
    df1 = pd.DataFrame({"X": X, "Y": Y})
    justmiss = df1[['X','Y']][df1.X.isnull()]
    notmiss = df1[['X','Y']][df1.X.notnull()]    
    df2 = notmiss.groupby('X',as_index=True)
    
    d3 = pd.DataFrame({},index=[])
    d3["COUNT"] = df2.count().Y
    d3["MIN_VALUE"] = df2.sum().Y.index
    d3["MAX_VALUE"] = d3["MIN_VALUE"]
    d3["EVENT"] = df2.sum().Y
    d3["NONEVENT"] = df2.count().Y - df2.sum().Y
    
    if len(justmiss.index) > 0:
        d4 = pd.DataFrame({'MIN_VALUE':np.nan},index=[0])
        d4["MAX_VALUE"] = np.nan
        d4["COUNT"] = justmiss.count().Y
        d4["EVENT"] = justmiss.sum().Y
        d4["NONEVENT"] = justmiss.count().Y - justmiss.sum().Y
        d3 = d3.append(d4,ignore_index=True)
    
    d3["EVENT_RATE"] = d3.EVENT/d3.COUNT
    d3["NON_EVENT_RATE"] = d3.NONEVENT/d3.COUNT
    d3["DIST_EVENT"] = d3.EVENT/d3.sum().EVENT
    d3["DIST_NON_EVENT"] = d3.NONEVENT/d3.sum().NONEVENT
    d3["WOE"] = np.log(d3.DIST_EVENT/d3.DIST_NON_EVENT)
    d3["IV"] = (d3.DIST_EVENT-d3.DIST_NON_EVENT)*np.log(d3.DIST_EVENT/d3.DIST_NON_EVENT)
    d3["VAR_NAME"] = "VAR"
    d3 = d3[['VAR_NAME','MIN_VALUE', 'MAX_VALUE', 'COUNT', 'EVENT', 'EVENT_RATE', 'NONEVENT', 'NON_EVENT_RATE', 'DIST_EVENT','DIST_NON_EVENT','WOE', 'IV']]      
    d3 = d3.replace([np.inf, -np.inf], 0)
    d3.IV = d3.IV.sum()
    d3 = d3.reset_index(drop=True)
    
    return(d3)

def data_vars(df1, Churn):
    
    stack = traceback.extract_stack()
    filename, lineno, function_name, code = stack[-2]
    vars_name = re.compile(r'\((.*?)\).*$').search(code).groups()[0]
    final = (re.findall(r"[\w']+", vars_name))[-1]
    
    x = df1.dtypes.index
    count = -1
    
    for i in x:
        if i.upper() not in (final.upper()):
            if np.issubdtype(df1[i], np.number) and len(Series.unique(df1[i])) > 2:
                conv = mono_bin(Churn, df1[i])
                conv["VAR_NAME"] = i
                count = count + 1
            else:
                conv = char_bin(Churn, df1[i])
                conv["VAR_NAME"] = i            
                count = count + 1
                
            if count == 0:
                iv_mydata = conv
            else:
                iv_mydata = iv_mydata.append(conv,ignore_index=True)
    
    iv = pd.DataFrame({'IV':iv_mydata.groupby('VAR_NAME').IV.max()})
    iv = iv.reset_index()
    return(iv_mydata,iv)
```

## Decile Plots and Kolmogorov Smirnov(KS) Statistic

```python
from collections import OrderedDict
import pandas as pd
import sys
%matplotlib inline
    
def plot_pandas_style(styler):
    from IPython.core.display import HTML
    html = '\n'.join([line.lstrip() for line in styler.render().split('\n')])
    return HTML(html)

def highlight_max(s,color='yellow'):
    '''
    highlight the maximum in a Series yellow.
    '''
    is_max = s == s.max()
    return ['background-color: {}'.format(color) if v else '' for v in is_max]

def decile_labels(agg1,label,color='skyblue'):
    agg_dummy = pd.DataFrame(OrderedDict((('TOTAL',0),('CHURN',0),('NONCHURN',0),('PCT_CHN',0),('CUM_CHN',0),('CUM_NONCHN',0),('DIST_CHN',0),('DIST_NONCHN',0),('SPREAD',0))),index=[0])
    agg1 = agg1.append(agg_dummy).sort_index()
    agg1.index.name = label
    agg1 = agg1.style.apply(highlight_max, color = 'yellow', subset=['SPREAD'])
    agg1.bar(subset=['CHURN'], color='{}'.format(color))
    agg1.bar(subset=['TOTAL'], color='{}'.format(color))
    agg1.bar(subset=['PCT_CHN'], color='{}'.format(color))
    return(agg1)

def deciling(data,decile_by,Churn,nonChurn):
    inputs = list(decile_by)
    inputs.extend((Churn,nonChurn))
    decile = data[inputs]
    grouped = decile.groupby(decile_by)
    agg1 = pd.DataFrame({},index=[])
    agg1['TOTAL'] = grouped.sum()[nonChurn] + grouped.sum()[Churn]
    agg1['CHURN'] = grouped.sum()[Churn]
    agg1['NONCHURN'] = grouped.sum()[nonChurn]
    agg1['PCT_CHN'] = grouped.mean()[Churn]*100
    agg1['CUM_CHN'] = grouped.sum()[Churn].cumsum()
    agg1['CUM_NONCHN'] = grouped.sum()[nonChurn].cumsum()
    agg1['DIST_CHN'] = agg1['CUM_CHN']/agg1['CHURN'].sum()*100
    agg1['DIST_NONCHN'] = agg1['CUM_NONCHN']/agg1['NONCHURN'].sum()*100
    agg1['SPREAD'] = (agg1['DIST_CHN'] - agg1['DIST_NONCHN'])
    agg1 = decile_labels(agg1,'DECILE',color='skyblue')
    return(plot_pandas_style(agg1))
```

```python
def scoring(features,clf,Churn):
    score = pd.DataFrame(clf.predict_proba(features)[:,1], columns = ['SCORE'])
    score['DECILE'] = pd.qcut(score['SCORE'].rank(method = 'first'),10,labels=range(10,0,-1))
    score['DECILE'] = score['DECILE'].astype(float)
    score['CHURN'] = Churn
    score['NONCHURN'] = 1 - Churn
    return(score)
```

```python
def gains(data,decile_by,Churn,score):
    inputs = list(decile_by)
    inputs.extend((Churn,score))
    decile = data[inputs]
    grouped = decile.groupby(decile_by)
    agg1 = pd.DataFrame({},index=[])
    agg1['ACTUAL'] = grouped.mean()[Churn]*100
    agg1['PRED'] = grouped.mean()[score]*100
    agg1['DIST_CHN'] = grouped.sum()[Churn].cumsum()/grouped.sum()[Churn].sum()*100
    agg1.index.name = 'DECILE'
    agg1 = agg1.reset_index()
    agg1['DECILE'] = agg1['DECILE']*10
    agg1['LIFT'] = agg1['DIST_CHN']/agg1['DECILE']
    plots(agg1,Churn,'Distribution')
```
